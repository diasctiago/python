```python
T = list(input())
C = list(input())
C = list(filter(lambda a: a != ' ', C))
def main():
    from sys import stdin, stdout

    tea = int(T[0])
    correct = 0
    for guess in list(C):
      if tea == int(guess):
        correct = correct + 1

         
    stdout.write(str(correct))
main()
```
