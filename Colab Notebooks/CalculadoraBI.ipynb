```
import pandas as pd
```

```
etl = pd.read_csv('etl.csv')
etl.head()
```

```
dataviz = pd.read_csv('dataviz.csv')
dataviz.head()
```

```
lista_tipo_etl = ['DIMENSÃO','DIMENSÃO','DIMENSÃO','FATO','FATO','FATO','FATO','ETL','AJUSTES','GESTAO','RISCO']

lista_tipo_dataviz = ['TELA','FILTRO','GRAFICO','TABELA','BOTÃO','AJUSTES','AJUSTES','AJUSTES','GESTAO','RISCO']
```

```
lista_atividade_etl = ['Modelagem multidimencional banco de dados',
                       'Desenvolvimento ferramenta de ETL',
                       'Prévia teste de cargas',
                       'Modelagem/Criação multidimencional banco de dados',
                       'Desenvolvimento ferramenta de ETL',
                       'Modelagem/Criação de campos calculados (regras de negócio)',
                       'Prévia teste de cargas',
                       'Criação de job para execução das cargas automatizadas',
                       'Ajustes pós homologação',
                       '10% do total para Gestão do Projeto',
                       '5% do total para Gestão de Risco']

lista_atividade_dataviz = ['Prototipagem e parte gráfica (Wireframe, cores, fontes, logotipo, etc.)',
                           'Filtro','Gráfico','Tabela','Botão','Publicação em homologação',
                           'Validação dos dados disponibilizados','Ajustes pós homologação',
                           '10% do total para Gestão do Projeto',
                           '5% do total para Gestão de Risco']
```

```
lista_horas_etl = [2,2,1,4,4,2,2,1,4,0,0]

lista_horas_dataviz = [4,1,4,6,1,1,2,4,0,0]
```

```
regras_etl = {'tipo': lista_tipo_etl,
              'atividades': lista_atividade_etl,
              'horas': lista_horas_etl}

regras_dataviz = {'tipo': lista_tipo_dataviz,
                  'atividades': lista_atividade_dataviz,
                  'horas': lista_horas_dataviz}
```

```
regras_etl = pd.DataFrame(regras_etl)
regras_etl
```

```
regras_dataviz = pd.DataFrame(regras_dataviz)
regras_dataviz
```

## Projeto ETL

```
df_etl = etl.join(regras_etl.set_index('tipo'), on='tipo')
df_etl['descricao'] = df_etl['nome'] + ' - ' + df_etl['atividades']
df_etl.head()
```

```
df_etl_ajustes = regras_etl.query('tipo in ("AJUSTES","ETL")').copy()
df_etl_ajustes['descricao'] = df_etl_ajustes['atividades']
df_etl_ajustes
```

```
df_etl_gestao = regras_etl.query('tipo in ("GESTAO")').copy()
df_etl_gestao['horas'] = df_etl['horas'].sum() * 0.1
df_etl_gestao['descricao'] = df_etl_gestao['atividades']
df_etl_gestao
```

```
df_etl_risco = regras_etl.query('tipo in ("RISCO")').copy()
df_etl_risco['horas'] = df_etl['horas'].sum() * 0.05
df_etl_risco['descricao'] = df_etl_risco['atividades']
df_etl_risco
```

```
projeto_etl = [df_etl, df_etl_ajustes, df_etl_gestao, df_etl_risco]
projeto_etl = pd.concat(projeto_etl)
projeto_etl.head()
```

```
print('######   Total de Horas do Projeto', round(projeto_etl['horas'].sum(), 2), '   ######')
projeto_etl[['descricao','horas']].head()
```

## Projeto DATAVIZ

```
df_dataviz = dataviz.join(regras_dataviz.set_index('tipo'), on='tipo')
df_dataviz['descricao'] = df_dataviz['nome']
df_dataviz
```

```
df_dataviz_tela = regras_dataviz.query('tipo in ("TELA")').copy()
df_dataviz_tela['horas'] = df_dataviz_tela['horas']*df_dataviz['tela'].value_counts().shape[0]
df_dataviz_tela['descricao'] = df_dataviz_tela['atividades']
df_dataviz_tela
```

```
df_dataviz_ajustes = regras_dataviz.query('tipo in ("AJUSTES")').copy()
df_dataviz_ajustes['descricao'] = df_dataviz_ajustes['atividades']
df_dataviz_ajustes
```

```
df_dataviz_gestao = regras_dataviz.query('tipo in ("GESTAO")').copy()
df_dataviz_gestao['horas'] = df_dataviz['horas'].sum() * 0.1
df_dataviz_gestao['descricao'] = df_dataviz_gestao['atividades']
df_dataviz_gestao
```

```
df_dataviz_risco = regras_dataviz.query('tipo in ("RISCO")').copy()
df_dataviz_risco['horas'] = df_dataviz['horas'].sum() * 0.05
df_dataviz_risco['descricao'] = df_dataviz_risco['atividades']
df_dataviz_risco
```

```
projeto_dataviz = [df_dataviz_tela, df_dataviz, df_dataviz_ajustes, df_dataviz_gestao, df_dataviz_risco]
projeto_dataviz = pd.concat(projeto_dataviz)
projeto_dataviz.head()
```

```
print('######   Total de Horas do Projeto', round(projeto_dataviz['horas'].sum(), 2), '   ######')
projeto_dataviz[['descricao','horas']]
```
