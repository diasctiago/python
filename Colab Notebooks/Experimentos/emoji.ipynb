## Transformando Textos com Emoji

```python
def emoji_transform(text):
  text_new = []
  for t in text.split():
    if emoji.is_emoji(t):
      for e in emoji.demojize(t, language='pt').replace(':','').split('_'):
        text_new.append(e)
    else:
      text_new.append(t)
  return ' '.join(map(str, text_new))
```

```python
emoji_transform('Python √© üëç')
```

## Alterando Textos com Emoji

```python
def emoji_replace(text, new_text):
  text_new = []
  for t in text.split():
    if emoji.is_emoji(t):
      text_new.append(new_text)
    else:
      text_new.append(t)
  return ' '.join(map(str, text_new))
```

```python
text = 'Pensa nas conversas das redes sociais, tipo: Eu ‚ù§Ô∏è o üé≤ ao 3Ô∏è‚É£ !!! Python √© üëç'
new_text = 'emoji'
emoji_replace(text, new_text)
```

## Contando Emojis em Textos

```python
def emoji_count(text):
  emoji_count = 0
  text_count = 0
  for t in text.split():
    if emoji.is_emoji(t):
      emoji_count += 1
    else:
      text_count += 1
  return (emoji_count, text_count)
```

```python
text = 'Pensa nas conversas das redes sociais, tipo: Eu ‚ù§Ô∏è o üé≤ ao 3Ô∏è‚É£ !!! Python √© üëç'
emoji_count(text)
```
