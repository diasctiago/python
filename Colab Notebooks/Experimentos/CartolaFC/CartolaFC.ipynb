```python
def get_dados(rodada):
  # RequisiÃ§Ã£o dos dados
  url        = 'https://api.cartola.globo.com/atletas/pontuados/'+ str(rodada)
  parametros = {}
  resposta   = requests.request("GET", url, params=parametros)
  try:
    objetos  = json.loads(resposta.text)
    return objetos
  except:
    return print('Erro rodada {}'.format(rodada))
```

```python
scouts_rodada = []
for r in range(atual_rodada):
  scouts_rodada.append(get_dados(r+1))
```

```python
scouts_rodada[0].keys()
```

```python
def get_partida(rodada):
  # RequisiÃ§Ã£o dos dados
  url        = 'https://api.cartola.globo.com/partidas/'+ str(rodada)
  parametros = {}
  resposta   = requests.request("GET", url, params=parametros)
  try:
    objetos  = json.loads(resposta.text)
    return objetos
  except:
    return print('Erro rodada {}'.format(rodada))
```

```python
partida_rodada = []
for r in range(atual_rodada+1):
  partida_rodada.append(get_partida(r+1))
```

```python
#partida_rodada[2].keys()
```

```python
#partida_rodada[2]
```

```python
def get_rodadas(rodada):
  df_rodadas = pd.json_normalize(partida_rodada[rodada]['partidas'])
  df_rodadas['rodada_id'] = partida_rodada[rodada]['rodada']
  df_clubes_ro = pd.json_normalize(partida_rodada[rodada]['clubes'].values())
  df_clubes_ro.rename(columns={'abreviacao': 'clube_visitante', 'id':'clube_visitante_id'}, inplace=True)
  df_rodadas = df_rodadas.join(df_clubes_ro[['clube_visitante_id','clube_visitante']].set_index('clube_visitante_id'), on='clube_visitante_id')
  df_clubes_ro = pd.json_normalize(partida_rodada[rodada]['clubes'].values())
  df_clubes_ro.rename(columns={'abreviacao': 'clube_casa', 'id':'clube_casa_id'}, inplace=True)
  df_rodadas = df_rodadas.join(df_clubes_ro[['clube_casa_id','clube_casa']].set_index('clube_casa_id'), on='clube_casa_id')
  return df_rodadas[['rodada_id','clube_casa_id','clube_casa','clube_visitante_id','clube_visitante']]
```

```python
df_rodadas_full = pd.DataFrame()
for r in range(atual_rodada+1):
  df_temp_ro = get_rodadas(r)
  df_rodadas_full = pd.concat([df_rodadas_full, df_temp_ro], ignore_index=True, sort=False)
```

## EscalaÃ§Ã£o

```python
df_rodada_clube_casa = df_rodadas_full.query('rodada_id == @atual_rodada+1').rename(columns={'clube_casa': 'clube','clube_visitante': 'clube_adversario'})[['clube','clube_adversario']]
df_rodada_clube_visi = df_rodadas_full.query('rodada_id == @atual_rodada+1').rename(columns={'clube_visitante': 'clube','clube_casa': 'clube_adversario'})[['clube','clube_adversario']]
```

```python
def anl_jogadoes(filtro_apelido,posicao,analise):
  df_escalacao = df[['apelido','clube','posicao','minimo_para_valorizar']].query('apelido in (@filtro_apelido)').drop_duplicates()
  if analise in ('ataque', 'meio'):
    df_anl_jogador = df[['apelido','clube','posicao','ATA','FIN','G','A','DS','FS']].groupby(by=['apelido','clube','posicao']).sum().reset_index()
    df_escalacao = pd.merge(df_escalacao, df_anl_jogador,  how='left', left_on=['apelido','clube','posicao'], right_on = ['apelido','clube','posicao'])
  elif analise in ('lateral', 'zaga'):
    df_anl_jogador = df[['apelido','clube','posicao','DEF','FIN','G','A','DS','FS','FC','CA','CV']].groupby(by=['apelido','clube','posicao']).sum().reset_index()
    df_escalacao = pd.merge(df_escalacao, df_anl_jogador,  how='left', left_on=['apelido','clube','posicao'], right_on = ['apelido','clube','posicao'])
  else:
    df_anl_jogador = df[['apelido','clube','posicao','PEG','DP','DE','GS','FS','CA','CV']].groupby(by=['apelido','clube','posicao']).sum().reset_index()
    df_escalacao = pd.merge(df_escalacao, df_anl_jogador,  how='left', left_on=['apelido','clube','posicao'], right_on = ['apelido','clube','posicao'])
  df_escalacao = pd.merge(df_escalacao, df_rodada_clube_casa,  how='left', left_on=['clube'], right_on = ['clube'])
  df_escalacao = pd.merge(df_escalacao, df_rodada_clube_visi,  how='left', left_on=['clube'], right_on = ['clube'])
  df_escalacao['clube_adversario'] = df_escalacao.clube_adversario_x.combine_first(df_escalacao.clube_adversario_y)
  df_escalacao.drop(columns=['clube_adversario_x','clube_adversario_y'],inplace=True)
  df_group_clube = df[['clube','G','DS']].groupby(by='clube').sum().rename(columns={'G': 'Gols Marcados','DS': 'Desarmes Feitos'})
  df_group_adversario =df[['clube_adversario','G','DS']].groupby(by='clube_adversario').sum().rename(columns={'G': 'Gols Sofridos','DS': 'Bolas Perdidas'})
  df_escalacao = pd.merge(df_escalacao, df_group_clube,  how='left', left_on=['clube_adversario'], right_on = ['clube'])
  df_escalacao = pd.merge(df_escalacao, df_group_adversario,  how='left', left_on=['clube_adversario'], right_on = ['clube_adversario'])
  if analise in ('ataque', 'meio'):
    return df_escalacao.query('posicao == @posicao').sort_values('ATA',ascending=False).reset_index(drop=True)
  elif analise in ('lateral', 'zaga'):
    return df_escalacao.query('posicao == @posicao').sort_values('DEF',ascending=False).reset_index(drop=True)   
  else:
    return df_escalacao.query('posicao == @posicao').sort_values('PEG',ascending=False).reset_index(drop=True)   
```

```python
def get_filtro_apelido(scout,n,posicao):
  df_filtro_apelido = df[['apelido','clube','posicao',scout]].query('posicao == @posicao').groupby(by=['apelido','clube','posicao']).sum().sort_values(scout,ascending=False).head(n)
  return df_filtro_apelido.index.get_level_values('apelido')
```
