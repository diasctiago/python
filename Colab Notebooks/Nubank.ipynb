```python
import string
```

```python
alphabet = list(string.ascii_uppercase)
alphabet_dict = {}
for a in range(len(alphabet)):
  alphabet_dict[alphabet[a]] = a + 1
#alphabet_dict
```

```python
alphabet_reverse = sorted(alphabet, reverse=True)
alphabet_reverse_dict = {}
for a in range(len(alphabet_reverse)):
  alphabet_reverse_dict[alphabet_reverse[a]] = a + 1
#alphabet_reverse_dict
```

```python
s = 'BZA'
list_string = list(s)
```

```python
time = 0
for i in range(len(list_string)-1):
  print()
  print(list_string[i])
  print(int(alphabet_dict[list_string[i]]))
  print(int(alphabet_reverse_dict[list_string[i]]))
  print(i+1)
  print(int(alphabet_dict[list_string[i+1]]))
  print(int(alphabet_reverse_dict[list_string[i+1]]))
  if list_string[i] == 'A':
    time = time + int(alphabet_reverse_dict[list_string[i+1]])
  elif list_string[i] < list_string[i+1]:
    time = time + int(alphabet_dict[list_string[i+1]]) - int(alphabet_dict[list_string[i]])
  else:
    time = time + int(alphabet_reverse_dict[list_string[i+1]]) - int(alphabet_reverse_dict[list_string[i]])

time
```

```python
import string

alphabet = string.ascii_lowercase 

word = 'VGXGPUAMKX'.lower()

distances = {l: min(abs(alphabet.index(l) - alphabet.index(c)), 26 - abs(alphabet.index(l) - alphabet.index(c))) for l, c in zip(word, 'a' + word[:-1])}
print(distances)
    
print(sum(distances.values()))
```
