### Opening the JSON files
### I need to improve this part and create a method to read several JSON files inside a path and loop all over them to json_normalize

```python
src = "/media/sf_data-science/json/tim/audiencias_customizadas/"
my_list = []

files = glob.glob('*', recursive=True)
```

```python
for single_file in files:
    with open(single_file, 'r') as f:
        json_file = json.load(f)
        my_list.append(json_normalize(data=json_file['data'], record_path='CUSTOMCLUSTER'))
```

### Inserting new columns of DATE to filter into Data Studio

```python
for file, i in zip(my_list,range(len(my_list))):
  file['channel'], file['date'] = [channels[i], current_date]
```
