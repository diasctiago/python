```
import pandas as pd
```

```
df = pd.read_csv('exemplo.csv')
```

```
df.head()
```

```
df.shape
```

```
lista_palavras = list(df['palavra'])
lista_palavras
```

```
lista_qtd = list(df['qtd'])
lista_qtd
```

```
lista_geral = []

for i in range(df.shape[0]):
  total_palavras = lista_qtd[i]*(','+lista_palavras[i])
  lista_total_palavras = total_palavras.split(sep=',')
  lista_total_palavras.remove('')
  lista_geral.append(lista_total_palavras)
```

```
lista_geral
```

```
lista_final = [item for sublista in lista_geral for item in sublista]
lista_final
```

```
with open('arquivo.txt', 'w') as arquivo:
    arquivo.write(', '.join(lista_final))
```

```
casos = int(input())
cont = int(0)
#completes os espaços vazios com sua solução
while (cont < casos):
    numero = int(input())
    if numero>0:
        if numero%2 == 0:
            print("EVEN POSITIVE")
        else:
            print("ODD POSITIVE")
    elif numero<0:
        if numero%2 == 0:
            print("EVEN NEGATIVE")
        else:
            print("ODD NEGATIVE")
    else:
        print("NULL")
    cont = cont + 1
```

```
T = int(input())
cont = int(0)
#completes os espaços vazios com sua solução
while (cont < T):
    S = input()
    resultado = 1
    for i in list(S):
      if i.upper() in ('A', 'E', 'I', 'O' , 'S'):
        resultado = resultado * 3
      else:
        resultado = resultado * 2
    print(resultado)
    cont = cont + 1
```
