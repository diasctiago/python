```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt #to display our wordcloud
from matplotlib.pyplot import figure
from PIL import Image #to load our image
import numpy as np #to get the color of our image
import pandas as pd
```

```python
df = pd.read_csv('table.csv', encoding = "iso-8859-1")
```

```python
df.head()
```

```python
df.shape
```

```python
word_list = list(df['words'])
#word_list
```

```python
list_qty = list(df['count'])
#list_qty
```

```python
list_qty[0] // 10
```

```python
general_list = []

for i in range(df.shape[0]):
  total_words = (list_qty[i] // 100) * (',' + word_list[i])
  total_words = total_words.split(sep = ',')
  total_words.remove('')
  general_list.append(total_words)
```

```python
#general_list
```

```python
final_list = [item for sublist in general_list for item in sublist]
#final_list
```

```python
todos_itens = ','.join(s for s in final_list)
```

```python
todos_itens
```

```python
with open('arquivo.txt', 'w') as file:
    file.write(', '.join(final_list))
```

```python
#Content-related
text = open('arquivo.txt', 'r').read()
stopwords = set(STOPWORDS)
```

```python
#Appearance-related
custom_mask = np.array(Image.open('Tiago2.JPG')) 
wc = WordCloud(background_color = 'white',
               stopwords = stopwords,
               #mask = custom_mask,
              # contour_width = 3,
              # contour_color = 'black'
              )
```

```python
wc.generate(text)
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func = image_colors)
```

```python
#Plotting
plt.figure(figsize = (10, 10))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()

#wc.to_file('my_export.png')
```

```python
# criar uma wordcloud
wc = WordCloud(stopwords=stopwords, background_color="black", width=1600, height=800)
wordcloud = wc.generate(todos_itens)

# plotar wordcloud
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
```
