```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL do site que qer pegar os dados
html = urlopen("https://receita.economia.gov.br/acesso-rapido/agenda-tributaria/agenda-tributaria-2020/agenda-tributaria-janeiro-2020/dia-06-01-2020")
bs = BeautifulSoup(html, 'html.parser')
```

```python
# Pegando os dados da tag TR
linhas = bs.find_all('tr', {'class':'even'})
```

```python
## Imprime todo texto contido em cada linha ##
for i in linhas:
    print(i.text)
```

```python
## Imprime o texto de cada uma das tags filhas ##
for i in linhas:
    filhas = i.findChildren("td")
    print(filhas[0])
    print(filhas[1])
    print(filhas[2])
```

```python
# Separando os dados em colunas
codigo, descricao, periodo = [], [], []
for i in linhas:
    children = i.findChildren("td")
    codigo.append(children[0].text)
    descricao.append(children[1].text)
    periodo.append(children[2].text)
```

```python
# Organizando os dados com o pandas
df = pd.DataFrame({'Código': codigo, 'Descrição': descricao, 'Período': periodo})
df.head()
```

```python
# Exportando o data frame
df.to_excel('arquivo.xlsx')
```
